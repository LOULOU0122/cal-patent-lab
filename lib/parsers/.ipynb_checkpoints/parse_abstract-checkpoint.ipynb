```python
import re
with open("patent_title_abstract_claims.tsv.00") as fp:
    for i, line in enumerate(fp):
        if i > 200:
            break
        temp = line.split('\t',1)
        matcher = re.compile('(?<=ABSTRACT.)(.*)(?=CLAIMS.)')
        patent_number = temp[0]
        #abstract = temp[1]
        #print temp[1]
        abstract = matcher.findall(temp[1])[0]
        if len(abstract) > 3:
            f = open(patent_number + ".txt", 'w')
            f.write(abstract)
            
```
