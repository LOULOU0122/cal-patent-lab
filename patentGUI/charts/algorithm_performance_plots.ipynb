```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')

# Simple function to prettify axes
# E.g.:
# ax = plt.subplot(111)
# simpleaxis(ax)
def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

```python
invalidation = [0.5, 0.62, 0.77]
denial = [0.5, 0.59, 0.69]

width = 0.25
pos = list(range(3))
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos,
        #using invalidation data,
        invalidation,
        # of width
        width,
        # with color
        color= '#9ecae1',
        # with label the first value in first_name
        label="Invalidation algorithm")

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos],
        #using df['mid_score'] data,
        denial,
        # of width
        width,
        # with color
        color='#253494',
        # with label the second value in first_name
        label="Denial algorithm")

# Set the y axis label
ax.set_ylabel('Accuracy')

# Set the chart's title
ax.set_title('Prediction algorithm accuracy scores')

# Set the position of the x ticks
ax.set_xticks([p + 1 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(['Random guessing', 'Baseline \n(due to data skew)', 'Logistic regression with \n stemming and n-grams'])
ax.set_yticklabels(["{}%".format(x) for x in [0, 20, 40, 60, 80, 100]])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, 1] )

# Prettify axis
simpleaxis(ax)

# Adding the legend and showing the plot
plt.legend(loc='upper left')

# plt.show()
plt.savefig('prediction_algorithm_accuracy.eps', format='eps', dpi=300)
plt.show()
```
